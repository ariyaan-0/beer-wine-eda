```python
from collections import defaultdict


SKU_list=df["SKU"].unique()

location_SKU_dict = defaultdict()

for location in df["Location"].unique():
    subdata= df[df["Location"]== location]
    sku_dict=defaultdict()
    
    sku_dict["total"]=len(subdata)
    
    for sku in SKU_list:
        sku_count = len(subdata[subdata["SKU"]==sku])
        sku_dict[sku]=sku_count
        
    for sku in SKU_list:
        sku_count = len(subdata[subdata["SKU"]==sku])
        sku_dict[sku+"_%"]=(sku_count/len(subdata))*100
        
    location_SKU_dict[location] = sku_dict
    
location_SKU_dict
        
```

```python
dict_to_df = pd.DataFrame.from_dict(location_SKU_dict, orient='index')

dict_to_df.reset_index(inplace=True)
dict_to_df.rename(columns={'index': 'Location'}, inplace=True)
```

## Predict Quality Score

```python
df=df.drop(['Brew_Date'], axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
categorical_features = df.select_dtypes(include=['object']).columns
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns
numerical_features = numerical_features[numerical_features != 'Quality_Score'] 
target_variable = 'Quality_Score'
```

```python
df = df.dropna(subset=[target_variable] + list(categorical_features) + list(numerical_features))
```

```python
X_train_numeric, X_test_numeric, y_train, y_test = train_test_split(df[numerical_features], df[target_variable], test_size=0.2, random_state=42)
X_train_cat, X_test_cat, _, _ = train_test_split(df[categorical_features], df[target_variable], test_size=0.2, random_state=42)
```

### Creating transformers for numerical and categorical features

```python
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combining transformers using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Creating and training the linear regression model within a pipeline
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
```

```python
model.fit(pd.concat([X_train_numeric, X_train_cat], axis=1), y_train)
```

```python
predictions = model.predict(pd.concat([X_test_numeric, X_test_cat], axis=1))


results_df = pd.DataFrame({'Actual': y_test, 'Predicted': predictions})


results_df.head(20)
```

```python
mse = mean_squared_error(y_test, predictions)
print(f'Mean Squared Error: {mse}')
```
